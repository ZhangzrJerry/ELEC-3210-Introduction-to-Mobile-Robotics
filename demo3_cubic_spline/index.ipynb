```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0,0.5], [0.5,3], [3,2], [6,5],[5,2], [8,3],[5,1],[8,1]])
n = len(x) - 1
m = len(x[0])
t1 = np.reshape(np.linspace(0, 1, 1001), [1, 1001])
t2 = t1 ** 2
t3 = t1 ** 3
t0 = np.reshape([1 for _ in range(1001)], [1, 1001])
tmp = np.zeros((n, n + 1))
tmp[0, 0] = 2
tmp[0, 1] = -3
tmp[0, 2] = 1
for i in range(1,n-1):
    tmp[i, i] = 1
    tmp[i, i+1] = -2
    tmp[i, i+2] = 1
tmp[-1, -1] = -1
tmp[-1, -2] = 1
A = tmp.dot(x)
tmp = np.zeros((n, n + 1))
tmp[0, 0] = -3
tmp[0, 1] = 4
tmp[0, 2]= -1
for i in range(1,n-1):
    tmp[i, i] = -1
    tmp[i, i+1] = 2
    tmp[i, i+2] = -1
tmp[-1, -1] = 1
tmp[-1, -2] = -1
B = tmp.dot(x)
tmp = np.zeros((n, n + 1))
for i in range(1, n):
    tmp[i, i] = -1
    tmp[i, i+1] = 1
C = tmp.dot(x)
tmp = np.zeros((n, n + 1))
for i in range(0, n):
    tmp[i, i] = 1
D = tmp.dot(x)
y = np.array([])
for i in range(n):
    if len(y) == 0:
        y = np.reshape(A[i], [-1, 1]).dot(t3) + \
            np.reshape(B[i], [-1, 1]).dot(t2) + \
            np.reshape(C[i], [-1, 1]).dot(t1) + \
            np.reshape(D[i], [-1, 1]).dot(t0)
    else:
        y = np.append(y, (
            np.reshape(A[i], [-1, 1]).dot(t3) + \
            np.reshape(B[i], [-1, 1]).dot(t2) + \
            np.reshape(C[i], [-1, 1]).dot(t1) + \
            np.reshape(D[i], [-1, 1]).dot(t0)
        ), axis=1)
plt.plot(y[0], y[1], color='red')
plt.scatter(x.T[0], x.T[1], color='blue')
plt.gca().set_aspect('equal', adjustable='box')
```
